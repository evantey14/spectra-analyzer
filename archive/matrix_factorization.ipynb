```python
from datetime import datetime
import numpy as np
import tables
import tensorflow as tf
import sys
```

```python
h5file = tables.open_file('sample.h5', 'r')
spectra = h5file.get_node("/spectra")[:]["spectrum"]
normalized_spectra = spectra / np.max(spectra, axis=1)[:, np.newaxis]
h5file.close()
```

```python
NUM_SPECTRA = 2000
NUM_FEATURES = 7
tf.reset_default_graph()

data_placeholder = tf.placeholder(tf.float32, shape=normalized_spectra[:NUM_SPECTRA, ::4].shape)
data = tf.Variable(data_placeholder, trainable=False)

a = tf.get_variable('a',
                    shape=(NUM_SPECTRA, NUM_FEATURES), 
                    dtype=tf.float32,
                    initializer=tf.orthogonal_initializer)
b = tf.get_variable('b',
                    shape=(NUM_FEATURES, 1569128//4), 
                    dtype=tf.float32,
                    initializer=tf.orthogonal_initializer)

product = tf.matmul(a, b)
squared_error = tf.reduce_mean(tf.square(tf.subtract(data, product)))

LEARNING_RATE = 0.01
global_step = tf.Variable(0, trainable=False)
optimize = tf.train.AdamOptimizer(LEARNING_RATE).minimize(squared_error, global_step=global_step)

NUM_ITERATIONS = 100000
PRINT_FREQ = 10
a_star = None
b_star = None
prev_se = sys.float_info.max
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer(), 
             feed_dict={data_placeholder: normalized_spectra[:NUM_SPECTRA, ::4]})
    for i in range(NUM_ITERATIONS):
        opt, se, a_star, b_star = sess.run([optimize, squared_error, a, b])
        if i % PRINT_FREQ == 0:
            print(i, '\t', se)
```

```python
# save info
outfile_name = 'output-{}'.format(datetime.now().strftime('%y%m%d-%H%M%S'))
metadata = np.array((NUM_SPECTRA, NUM_FEATURES, LEARNING_RATE))
np.savez_compressed(outfile_name, metadata=metadata, a_star=a_star, b_star=b_star)
```
